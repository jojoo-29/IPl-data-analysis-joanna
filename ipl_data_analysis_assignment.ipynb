# IPL Data Analysis
**Assignment 1 – IPL Data Analysis**

This notebook analyzes IPL datasets (`matches.csv` and `deliveries.csv`) to generate insights about match outcomes, team performance, player performance, and venue trends.

## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
```

## 2. Load Dataset

```python
matches = pd.read_csv('matches.csv')
deliveries = pd.read_csv('deliveries.csv')

matches.head()
```

```python
deliveries.head()
```

## 3. Match Won by Maximum Runs

```python
max_runs = matches.loc[matches['win_by_runs'].idxmax()]
max_runs
```

**Insight:** This match had the highest run margin victory in IPL history.

## 4. Match Won by Maximum Wickets

```python
max_wickets = matches.loc[matches['win_by_wickets'].idxmax()]
max_wickets
```

**Insight:** The winning team chased the target very easily with the highest wicket margin.

## 5. Minimum Run Victory

```python
min_runs = matches[matches['win_by_runs'] > 0]['win_by_runs'].min()
matches[matches['win_by_runs'] == min_runs]
```

**Insight:** These matches were extremely close with the smallest run margin.

## 6. Minimum Wicket Victory

```python
min_wickets = matches[matches['win_by_wickets'] > 0]['win_by_wickets'].min()
matches[matches['win_by_wickets'] == min_wickets]
```

## 7. D/L Method Applied vs Not Applied

```python
sns.countplot(x='dl_applied', data=matches)
plt.title('D/L Method Applied vs Not Applied')
plt.show()
```

## 8. Matches Held in Each City

```python
city_matches = matches['city'].value_counts()
city_matches.head(10).plot(kind='bar')
plt.title('Top Cities Hosting IPL Matches')
plt.show()
```

## 9. Matches Won by Each Team

```python
team_wins = matches['winner'].value_counts()
team_wins.plot(kind='bar')
plt.title('Matches Won by Each Team')
plt.show()
```

## 10. Matches Played Each Season

```python
season_matches = matches['season'].value_counts().sort_index()
season_matches.plot(kind='bar')
plt.title('Matches Played Per Season')
plt.show()
```

## 11. Top 10 Man of the Match Players

```python
top_players = matches['player_of_match'].value_counts().head(10)
top_players.plot(kind='bar')
plt.title('Top 10 Man of the Match Winners')
plt.show()
```

## 12. Toss Winner vs Match Winner

```python
toss_match = matches[matches['toss_winner'] == matches['winner']]
percentage = len(toss_match) / len(matches) * 100
percentage
```

**Insight:** Winning the toss provides some advantage but does not guarantee match victory.

## 13. Toss Decision Analysis

```python
sns.countplot(x='toss_decision', data=matches)
plt.title('Toss Decisions (Bat vs Field)')
plt.show()
```

## 14. D/L Method Used Per Season

```python
dl_season = matches.groupby('season')['dl_applied'].sum()
dl_season.plot(kind='bar')
plt.title('D/L Method Usage Per Season')
plt.show()
```

## 15. Matches Played in Different Stadiums

```python
matches['venue'].value_counts().head(10).plot(kind='bar')
plt.title('Top Stadiums Hosting IPL Matches')
plt.show()
```

## 16. Top Run Scorers (Bonus Analysis)

```python
top_batsmen = deliveries.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
top_batsmen.plot(kind='bar')
plt.title('Top Run Scorers in IPL')
plt.show()
```

## Conclusion
- IPL matches show strong competition between teams.
- Some teams consistently perform better across seasons.
- Winning the toss gives a slight strategic advantage.
- Player performance such as Man of the Match awards highlights key contributors.
